```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=10
x=np.zeros(n)
y=np.zeros(n)
alpha=0.1
x[0]=1.0
for i in range(n-1):
    x[i+1]=x[i]-alpha*(2*x[i])

y=x**2
m=1.5
plt.plot(np.arange(-m,m,0.1),np.arange(-m,m,0.1)**2)
plt.plot(x,y,'or--')
```
