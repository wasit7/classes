## Note

### Gradient Descent

$
\begin{eqnarray*}
\beta_{i+1}&\Leftarrow&\beta_{i}-\alpha \frac {\partial{J}} {\partial \beta}\\
\end{eqnarray*}
$

### Sigmoid function

$
\begin{eqnarray*}
a(z)&=&\frac{1}{1+e^{-z}}\\
\frac {d{a}} {d z}&=&a(1-a)
\end{eqnarray*}
$

### Hyperbolic Tan function

$
\begin{eqnarray*}
a(z)&=&{tanh(z)}\\
\frac {d{a}} {d z}&=&1-a^2\\
\end{eqnarray*}
$
