# Basic Math

```python
x=np.arange(0,2*np.pi,0.1)
```

```python
y=np.sin(x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y,'ko')
```

```python
y_sin=np.sin(x)
y_cos=np.cos(x)
y_tanh=np.tanh(x)
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.plot(x,y_tanh)
```
