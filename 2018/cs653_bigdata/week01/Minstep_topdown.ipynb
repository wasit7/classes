```python
import sys
sys.setrecursionlimit(100000)
```

```python
m={1:0}
def minstep(x):
    #print("%s"%x)
    global m
    if x==1:
        return 0 #terminating case
    if x in m:
        return m[x] #memorization
    mins=minstep(x-1)+1
    if x%3==0:
        new_mins=minstep(x//3)+1
        if new_mins<mins:
            mins=new_mins
    if x%2==0:
        new_mins=minstep(x//2)+1
        if new_mins<mins:
            mins=new_mins
    m[x]=mins
    return mins
```
