```python
m={1:0}
for i in range(2,10000001,1):
    mins=m[i-1]+1
    if i%2==0:
        new_mins=m[i//2]+1
        if new_mins<mins:
            mins=new_mins
    if i%3==0:
        new_mins=m[i//3]+1
        if new_mins<mins:
            mins=new_mins
    m[i]=mins
```

```python
m={
    1:{
        'minstep': 0,
        'operation': 'None',
        'target': 'None'
    }
}
for i in range(2,1000001,1):
    m[i]={
        'minstep': 0,
        'operation': 'None',
        'target': 'None'
    }
    mins=m[i-1]['minstep']+1
    op='-1'
    target=i-1
    if i%2==0:
        new_mins=m[i//2]['minstep']+1
        if new_mins<mins:
            mins=new_mins
            op='//2'
            target=i//2
    if i%3==0:
        new_mins=m[i//3]['minstep']+1
        if new_mins<mins:
            mins=new_mins
            op='//3'
            target=i//3
    m[i]['minstep']=mins
    m[i]['operation']=op
    m[i]['target']=target
```
